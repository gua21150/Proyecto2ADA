```python
#librerias
import matplotlib.pyplot as plt

import time
## Funcion que dibuja la triangulación
## Esta funcion no es muy importante, solo ayuda a visualizar el codigo
def draw_polygon_with_triangulation(points, triangulation):
    ##Figura original
    polygon = points + [points[0]]
    xs, ys = zip(*polygon)
    plt.figure()
    plt.fill(xs, ys, 'lightgray', edgecolor='black', linewidth=1)
    plt.plot(xs, ys, marker='o')

    #Triangulación
    n = len(points)
    for i, j, k in triangulation:
        tri_xs, tri_ys = zip(*[points[i], points[j], points[k], points[i]])
        plt.plot(tri_xs, tri_ys, 'r--')  # Red dashed line for triangulation

    plt.show()


#Triangulación
def polygon_triangulation(points):
    start_time = time.time() ##O(1), empieza el timer
    n = len(points)          ##O(1), se define n, que ayuda a saber la cantidad de lados que tiene el polinomio

    def dist(p1, p2): 
        return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5 ##O(1), define una función que mide la distancia entre 2 vertices.
                                                              ##La funcion es fundamental para que la optimización sea óptima,
                                                              ##porque busca que sea mínima. 
            
    dp = [[0 for _ in range(n)] for _ in range(n)] #O(1) Crea una matriz, que aun no almacena nada pero aquì se guardaran los datos y costos de las aristas
    choice = [[-1 for _ in range(n)] for _ in range(n)] #O(1), otra matriz, la cual almacena las decisiones que se toman.

    for gap in range(2, n): abs ##O(n), itera sobre cada posible poligono que puede ser creado
        for i in range(n - gap): #O(n), itera sobre cada posible vertice donde puede empezar una triangulacion
            j = i + gap          #O(1), calcula el vertice donde 2 otros vertices llevarán su aristas, creando un triangulo.
            dp[i][j] = float('inf') #O(1), costo minimo 
            for k in range(i + 1, j): #O(n) Representa los vertices de en medio que pueden ser creados. 
                cost = dp[i][k] + dp[k][j] + dist(points[i], points[j]) + dist(points[i], points[k]) + dist(points[k], points[j]) #O(1) Calcula costo
                if dp[i][j] > cost: #O(1),agrega el costo a la matriz dp y la elección a la matriz choice
                    dp[i][j] = cost  
                    choice[i][j] = k

    def reconstruct_triangulation(i, j, choice, triangulation): # 4* O(1), son funciones ya establecidas, solo corren lo que ya esta guardado
        if choice[i][j] == -1:
            return
        k = choice[i][j]
        triangulation.append((i, k, j))
        reconstruct_triangulation(i, k, choice, triangulation)
        reconstruct_triangulation(k, j, choice, triangulation)

    triangulation = []
    reconstruct_triangulation(0, n-1, choice, triangulation)
    end_time = time.time()
    tiempo = end_time - start_time
    return dp[0][n-1], triangulation, tiempo

min_cost, triangulation, tiempo= polygon_triangulation(points)
print(f"Minimum triangulation cost: {min_cost}")
draw_polygon_with_triangulation(points, triangulation)


# Place the triangulation function call or any other operation here
min_cost, dp, tiempo = polygon_triangulation(points)
print("Tiempo de ejecución: ", tiempo)
```
